```python
import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
dataset = pd.read_csv('diabetes-dataset.csv')
dataset.head()
```

```python
dataset.shape
```

```python
feature_names = dataset.columns
print(feature_names)

x = dataset.drop(["Outcome"],axis=1)
y=dataset["Outcome"]

X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size = 0.3, random_state=6)
```

```python
Y_test.head()
```

```python
# Convert into float type
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

# Converting Y to categorical.
Y_train = to_categorical(Y_train, 2)
Y_test = to_categorical(Y_test, 2)

X_train.info()
X_train.head()
```

# First Model with one hidden layer with 16 neurons.

### Creating Model

```python
model_1 = Sequential()
model_1.add(Dense(8, input_shape=(8,), activation='relu'))
model_1.add(Dense(16, activation='relu'))
model_1.add(Dense(8, activation='relu'))
model_1.add(Dense(2, activation='softmax'))
```

### Configure the model and start training

```python
model_1.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

### Training the model

```python
model_1.fit(X_train, Y_train, epochs=1000, batch_size=250, verbose=1, validation_split=0.2)
```

### Predicting

```python
# Test the model after training
model_1.evaluate(X_test, Y_test, verbose=1)
```

We can see that our first model gave us validation accuracy of 83.50% which is very good.

# 4th Model with three hidden layer with 16, 32 and 64 neurons.

### Creating Model

```python
model_4 = Sequential()
model_4.add(Dense(8, input_shape=(8,), activation='relu'))
model_4.add(Dense(16, activation='relu'))
model_4.add(Dense(32, activation='relu'))
model_4.add(Dense(64, activation='relu'))
model_4.add(Dense(2, activation='softmax'))
```

```python
optimizers = ["sgd",
              "rmsprop",
              "adam",
              "adadelta",
              "adagrad",
              "adamax",
              "nadam",
              "ftrl"]
```

### Model configuration, training and prediction using different optimizers.

```python
for op in optimizers:
    
    print("Optimizer:",op)
    # Configure the model and start training
    model_4.compile(loss='categorical_crossentropy', optimizer=op, metrics=['accuracy'])
    
    # Fitting Model
    model_4.fit(X_train, Y_train, epochs=1000, batch_size=250, verbose=0, validation_split=0.2)


    # Test the model after training
    model_4.evaluate(X_test, Y_test, verbose=1)
    print("------------------------------------------------------------------------")
```

We can cleary see that adadela, adagrad, adamax and nadam outperform all other optimizers and adamax and nadam has the highest accuracy.
