# Apriori | FP-Growth | FP-Max

#### Apriori: 
It is a popular algorithms for frequent item set generation which are then used for producing mining rules. Apriori taes large space and more, is less accurate and take more time when dealing with large datasets.

#### FP-Growth: 
It stands for Frequent Pattern Growth and is a frequent pattern production algorithm that developes a pattern tree with items in it. Unlike, Aproiri, it increases linearly. It is more accurate and requires less memory and space.

#### FP-Max:
It is an extention of FP-Growth which tries to obtains maximal itemsets.
An particular itemset is considered maximal if it is frequent and there does not exist any super-pattern that contains that particular itemset.  



```python
dataset = [['Milk', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Dill', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Milk', 'Apple', 'Kidney Beans', 'Eggs'],
           ['Milk', 'Unicorn', 'Corn', 'Kidney Beans', 'Yogurt'],
           ['Corn', 'Onion', 'Onion', 'Kidney Beans', 'Ice cream', 'Eggs']]
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
from mlxtend.frequent_patterns import fpmax

fpmax(df, min_support=0.6)
```

```python
fpmax(df, min_support=0.6, use_colnames=True)
```
